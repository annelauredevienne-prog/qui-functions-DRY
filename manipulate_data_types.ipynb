```python
import requests
request = requests.get("https://swapi.py4e.com/api/people/5")
dic = request.json()

if dic["gender"] == "male" : 
    print ("{} is {} tall and weighs {}kg. He was born in {}".format(dic ["name"], dic["height"], dic["mass"], dic["birth_year"]))
else : 
    print ("{} is {} tall and weighs {}kg. She was born in {}".format(dic ["name"], dic["height"], dic["mass"], dic["birth_year"]))
```

```python
kilometres_travelled = [50, 10, 100, 25, 1000, 21, 12,30,60]
total_km = 0
total_users = 0

kilometres_travelled.pop(4)
total_km = 0
total_users = 0

for value in kilometres_travelled:
    total_km += value
    total_users += 1

average_km = total_km / total_users
print("On average, users travelled {} km to get to work.".format(average_km))
```
