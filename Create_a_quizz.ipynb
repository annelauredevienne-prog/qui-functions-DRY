```python

# Welcome message
print("Here you have our quiz !")

correct_answer = "Elon Musk"
max_attempts = 3

# 3 questions defined with always same anwser presented as list
questions = [
    ( "Who is the founder of Tesla?"),
    ("Which entrepreneur is famous with Tesla?"),
    ("Who is the director of SpacX?")
]

# Define the function calling all questions as a quiz
def ask_question(question):

    attempts = 0

    while attempts < max_attempts:

        answer = input(question)
        
        if answer == correct_answer: 
            print("Good job! This is the right answer")
            return

        attempts += 1

        if attempts < max_attempts:
            print(f"Incorrect answer. You have {max_attempts - attempts} attempts left.")
        
        else:
            print("Too bad! That is not the correct answer")

# ask 3 questions - boucle
for question in questions:
    ask_question(question)


 
```
