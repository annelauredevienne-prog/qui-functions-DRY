```python
# Define the correct password for Jedha
correct_password = "JEDHA"

# Initialize the loop. Many attemps can be made to try password. IT means there is a loop.
password_attempt = ""
attempts = 0
max_attempts = 3

# Welcome message
print("Welcome to JEDHA Authentication")

# Messages to attempt password
while password_attempt != correct_password and attempts < max_attempts:
    password_attempt = input("Please enter the password: ")
    attempts += 1

    if password_attempt == correct_password:
        print("Access granted! Welcome to JEDHA.")
    elif attempts < max_attempts:
        print("Incorrect password. You have {max_attempts - attempts} attempts left.")
    else:
        print("Too many failed attempts. Access denied.")

if attempts >= max_attempts and password_attempt != correct_password:
    print("Too many failed attempts. Access denied.")
```
