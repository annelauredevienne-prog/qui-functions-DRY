7. What is the most ordered item?

*Hint* 🤓
* Use `.groupby()` with a relevant aggregation function
* Then you can use `.sort_values()` to sort the results

```python
#regrouper par nom de produits
grouped = dataset.groupby("item_name")

#selectionne les quantités commandées par produit
quantity = grouped["quantity"]

#fais la somme des quantités commandées par produit
total = quantity.sum()

#classe les quantités par ordre descendant
sorted_total = total.sort_values(ascending=False)

print(sorted_total)
```

```python
#convert item_price to a decimal number

print(dataset["item_price"].dtype)
print(dataset["item_price"].head())
```

```python
dataset["revenue"] = dataset["item_price"] * dataset["quantity"]
total_revenue = dataset["revenue"].sum()
print(total_revenue)
```

9. What is the average revenue per order?

```python
#colonne avec les revenus par ligne
dataset["revenue"] = dataset["quantity"] * dataset["item_price"]

#colonne utiles
df = dataset [["order_id", "revenue"]] 

#regrouper par commande
grouped = df.groupby("order_id")

#revenue pour chaque commande
revenue_per_order = grouped["revenue"].sum()

#moyenne de revenus par commande 
revenue_per_order.mean()

print (revenue_per_order.mean())
```
